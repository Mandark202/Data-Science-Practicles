```python
import pandas as pd

matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

matches.head()
```

```python
wins = matches['winner'].value_counts()

wins.head(10)
```

```python
import matplotlib.pyplot as plt

wins.head(10).plot(kind='bar')
plt.title("Top Winning Teams in IPL")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.xticks(rotation=90)
plt.show()
```

```python
Mumbai Indians and Chennai Super Kings are among top winning teams
These teams dominate IPL history
```
