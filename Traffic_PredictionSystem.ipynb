```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("traffic_data.csv")

# Convert time
df["time"] = df["time"].str.replace(":", "").astype(int)

# Plot
df.groupby("time")["traffic_volume"].mean().plot(kind='bar')
plt.title("Traffic Volume by Time")
plt.xlabel("Time")
plt.ylabel("Number of Vehicles")
plt.show()
```

```python
df.groupby("weather")["traffic_volume"].mean().plot(kind='bar')
plt.title("Traffic Volume by Weather")
plt.xlabel("Weather")
plt.ylabel("Traffic Volume")
plt.show()
```

```python
df.groupby("weather")["traffic_volume"].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Traffic Distribution by Weather")
plt.ylabel("")
plt.show()
```

```python
df["date"] = pd.to_datetime(df["date"])

df.groupby("date")["traffic_volume"].sum().plot()
plt.title("Daily Traffic Trend")
plt.xlabel("Date")
plt.ylabel("Traffic Volume")
plt.show()
```
